```python
import json
harmbench_path = "../assets/harmbench/splits"
```

```python
import os
files = os.listdir(harmbench_path)
print(files)
```

```python
harmful_files = [f for f in files if "harmful" in f]
print(harmful_files)
```

```python
harmless_files = [f for f in files if "harmless" in f]
print(harmless_files)
```

```python
files_path = os.path.join(*harmbench_path.split("/")[:-1])
print(files_path)
```

```python
harmful_json_data = []
for file in harmful_files:
    with open(os.path.join(harmbench_path, file), "r") as f:
        harmful_json_data.append(json.load(f))
with open(os.path.join(files_path, "harmful_data.jsonl"), "w") as f:
    for line in harmful_json_data[0]:
        f.write(json.dumps(line) + "\n")
print(len(harmful_json_data[0]))
```

```python
harmless_json_data = []
for file in harmless_files:
    with open(os.path.join(harmbench_path, file), "r") as f:
        harmless_json_data.append(json.load(f))
print(len(harmless_json_data[0]))
with open(os.path.join(files_path, "harmless_data.jsonl"), "w") as f:
    for line in harmless_json_data[0]:
        f.write(json.dumps(line) + "\n")
```

```python
import datasets
harmful_dataset_path = os.path.join(files_path, "harmful_data.jsonl")
harmless_dataset_path = os.path.join(files_path, "harmless_data.jsonl")
harmful_dataset = datasets.load_dataset("json", data_files=harmful_dataset_path)["train"]
harmless_dataset = datasets.load_dataset("json", data_files=harmless_dataset_path)["train"]
print(len(harmful_dataset))
print(len(harmless_dataset))
```
